#4. Как вы думаете, какие еще признаки могут повлиять на итоговую оценку? Найдите закономерности, аналогичные рассмотренным

Рассмотрим влияние существования романтических отношений на итоговую оценку.
Верно ли, что студенты, имеющие отношения, учатся хуже, чем остальные?

```python
romantic = data[data['romantic'] == "yes"]
not_romantic =  data[data['absences'] == "no"]

romantic_g3 = romantic['G3'].mean()
not_romantic_g3 = not_romantic['G3'].mean()

if (romantic_g3 < not_romantic_g3 ):
    print("It's true")
else:
    print("It's false")
```

Рассмотрим влияние наличия домашнего интернета на итоговую оценку.
Верно ли, что студенты, имеющие интернет, учатся лучше, чем остальные?

```python
internet = data[data['internet'] == "yes"]
no_internet =  data[data['internet'] == "no"]

internet_g3 = internet['G3'].mean()
no_internet_g3 = no_internet['G3'].mean()

if (internet_g3 > no_internet_g3 ):
    print("It's true")
else:
    print("It's false")
```
